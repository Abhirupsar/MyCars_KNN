## Import numpy and pandas and also use read the cars dataset

```python
##import numpy and pandas library to perform necessary numeric and dataframe operations on the data

import numpy as np
import pandas as pd
import os

os.chdir("C:\\Users\Abhirup Sarkar\Downloads")

data = pd.read_csv("cars.csv")

data2 = data[(data["MPG"] > 0) & (data["Horsepower"] > 0)]
data2
```

```python
## Resetting the index of the dataset after the zero values were removed from the dataset
## Also separating the required columns

data_final = data2[["Car", "Cylinders", "Weight", "Horsepower", "MPG"]]
data_final.reset_index(drop=True, inplace=True)
data_final
```

## Calculating the scaled values of all variables and updating them as new columns in the Dataframe

```python
## We are adding scaled values of the variables as new columns in the data frame

data_final["Cylinders_s"] = (data_final["Cylinders"]- data_final["Cylinders"].mean())/ data_final["Cylinders"].std()
data_final["Weight_s"] = (data_final["Weight"]- data_final["Weight"].mean())/ data_final["Weight"].std()
data_final["Horsepower_s"] = (data_final["Horsepower"]- data_final["Horsepower"].mean())/ data_final["Horsepower"].std()
data_final
```

## Separating test and train datasets

```python
## Separating test data from the dataframe
## I have chosen to take 5 rows as test data to confirm the made KNN model by passing it through different data

# Here we are taking last 5 variables as test datas
df_test = data_final[-5:]   
df_test
```

```python
## Separating train data from the dataframe

df_train = data_final[:-5]  
df_train
```

## Function to calculate eucledian_distance and add it as a separate column in the train data set

```python
## This is the function to calculate eucledian distance for the variables
## y_test corresponds to Weight of the car from test dataset
## x_test corresponds to the number of cylinders from the test dataset
## z_test corresponds to the horsepower of the engine from the test dataset

def eucledian_distance(x_test, y_test, z_test, df_train ):

    df_train["Ecdn_dist"] = np.sqrt((float(y_test)-df_train["Weight_s"])**2 + 
                                    (float(z_test)- df_train["Horsepower_s"])**2 + 
                                    (float(x_test)- df_train["Cylinders_s"])**2 )
    
    #return[pd.DataFrame(np.transpose(df_train["Ecdn_dist"]))]
    return(df_train)
```

## Function to return nearerst K entries to the test data

```python
## This function will return nearest K entries to the test data in the form of a DataFrame

def knn(Data, test_data , k = 3, row_num=0):
    
    
    # Pass the needed row test values from the test dataframe into the necessary variables
    
    x_test = test_data['Cylinders_s'][row_num:row_num+1]
    y_test = test_data['Weight_s'][row_num:row_num+1]
    z_test = test_data['Horsepower_s'][row_num:row_num+1]
    
    
    #Below command will return a dataframe with newly added computed eucledian distance column
    Data = eucledian_distance(x_test, y_test, z_test, Data )
    
    
    '''Below code line will create a new dataframe which will have the above maodified dataframe sorted in 
    the form of increasing eucledian distance
    '''
    
    sorted_data = Data.sort_values(by= "Ecdn_dist",axis=0)
    
    
    '''
     This below code creates an empty dataframe. This dataframe is gonna be looped to contain the nearest 
     K-neighbours to the test dataset and this dataset will be returned as a result of this 
     function getting executed
     '''
    
    sorted_data2 = pd.DataFrame(columns =['Car', 'Cylinders', 'Weight', 'Horsepower', 'MPG', 'Cylinders_s',
                                       'Weight_s', 'Horsepower_s', 'Ecdn_dist'],
                                    dtype='object')
    
    ## Also in case there are same values in the K-nearest eucledian distances, we will take the next unique value 
    ##under consideration. The below code supports this.

    for i in sorted_data["Ecdn_dist"].unique()[:k]:
        
        y = Data[sorted_data["Ecdn_dist"]== i]
        sorted_data2 = sorted_data2.append(y,ignore_index=True)
    
    
    return(sorted_data2)
    #print(sorted_data2)
    #return((sorted_data2['MPG']).mean())
```

## Return the KNN values. By default the K value is kept at 3 in the function. It can be modified into other values as per the requirement

```python
## Executing the KNN function with k=2 and below mentioned parameters

knn(Data= df_train, test_data = df_test, k=2,row_num=4)
```

## Calculating the appropriate output through the KNN values

```python
## since this is a regression model, we are gonna return the mean of the K-values as the k_labels as the output

print("The appropriate value according to KNN model is: ")
knn(Data= df_train, test_data = df_test, k=2)['MPG'].mean()
```
